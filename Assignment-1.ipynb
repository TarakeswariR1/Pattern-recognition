$\;$Consider the following probability density function which is non-zero only in the range $$0 ≤ x ≤ 10$$ Here, K is a constant. Determine the value of the constant K.

$$p(x) = K*x^3(10 − x)$$ 

X is a continous random variable, with the above given probability density function. Since $$p(x)$$ is equal to 1 because  probability is between $$0 \leq x \leq 1$$

$$p(x) = \int_{0}^{10} Kx^3(10-x)dx$$
$$1 = \int_{0}^{10} Kx^3(10-x)dx$$
$$1=K \int_{0}^{10} (10x^3-x^4) dx$$

Integrating the above integral with respect to x we get.

$$1=K \big[\frac{5x^4}{2} - \frac{x^5}{5}] |_{0}^{10}$$
$$1=K \big[\frac{5(10)^4}{2} - \frac{(10)^5}{5}] - 0$$
$$1=K \big[25000 - 20000]$$
$$1=K 5000$$
$$\frac{1}{5000}=K$$
