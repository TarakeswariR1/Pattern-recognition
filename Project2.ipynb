```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics as st
from scipy.stats import laplace
from sklearn.datasets import make_blobs
from scipy.stats import norm
from numpy import mean
from numpy import std

trainDF=pd.read_csv("/Users/tarakeswariramachandra/Documents/Michigan State University/CSE 802/archive/sign_mnist_train.csv")


def ClassConditional(data):
    meann=mean(data)
    var=std(data)
    #print((meann, var))
    #dist = norm(meann, var)
	
    return (meann, var) if not None else 0


meanClass=[]
varrClass=[]
prior=[]
for i in range(1, 25):
    trainDfClass1=np.array(trainDF[trainDF["label"]==i])
    m, v=ClassConditional(trainDfClass1)
    p=len(trainDfClass1)/len(trainDF)
    meanClass.append(m)
    varrClass.append(v)
    prior.append(p)
prior.pop(8)
meanClass.pop(8)
varrClass.pop(8)
varrClass
```
